```python
from kafka import KafkaConsumer
import json
import pandas as pd
from sqlalchemy import create_engine
```

```python
consumer = KafkaConsumer(
    'plot-data',
    bootstrap_servers=['localhost:9092'],
    auto_offset_reset='earliest',
    enable_auto_commit=True,
    group_id='consumer-plot',
    value_deserializer=lambda x: json.loads(x.decode('utf-8'))
)
```

```python
engine = create_engine('sqlite:///sensor_data.db')
```

```python
for message in consumer:
    data = message.value
    
    df = pd.DataFrame([data], columns=["sensor_id", "temperature", "timestamp"])
    

    df.to_sql('sensor_readings', con=engine, if_exists='append', index=False)
    
    print(f"Received and saved to SQL: {data}")
```
